```python
class AliasSampler(object):
    def __init__(self, p):
        self.p = p
        self.U = np.zeros(len(p))
        self.K = np.zeros(len(p), dtype=np.int)
        self.generate(self.p)
    
    # generate
    def generate(self, p):
        
        smaller, bigger = [], []
        n = len(p)
        
        self.K = [n * pi for pi in p]
        print(self.K)
        
        for ind, item in enumerate(self.K):
            if item < 1:
                smaller.append(ind)
            else:
                bigger.append(ind)
        
        while smaller and bigger:
            small_ind = smaller[0]
            big_ind = bigger[0]
            self.U[small_ind] = big_ind
            self.K[big_ind] = self.K[big_ind] - (1. - self.K[small_ind])
            if self.K[big_ind] < 1.:
                bigger.pop(0)
                smaller.append(big_ind)
            smaller.pop(0)
        return

    def sample(self):
        if self.U is None or self.K is None:
            self.generate(self.p)
        
        n = len(self.p)
        
        x = np.random.rand()
        
        i = int(np.floor(n * x))
        
        if np.random.rand() < self.K[i]:
            return i
        else:
            return self.U[i]
         
```

```python
phi = [8., 3., 6., 11.]
pi = [x / sum(phi) for x in phi]
```

```python
plt.bar([_ for _ in range(len(pi))], pi)
```

```python
cs = np.cumsum(pi)
res = []
for _ in range(1000):
    draw = np.random.rand()
    res.append(sum(cs < draw))
```

```python
sns.distplot(res)
```

```python
Sampler = AliasSampler(pi)
print(Sampler.U)
print(Sampler.K)
```

```python
draws = []
for _ in range(10000):
    draw = Sampler.sample()
    draws.append(draw)
```

```python
sns.distplot(draws)
```
