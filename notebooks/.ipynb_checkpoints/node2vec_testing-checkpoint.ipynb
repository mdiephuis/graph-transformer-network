```python
# generate
def alias_generate(p):

    smaller, bigger = [], []
    n = len(p)

    K = [n * pi for pi in p]
    U = [0] * n

    for ind, item in enumerate(K):
        if item < 1:
            smaller.append(ind)
        else:
            bigger.append(ind)

    while smaller and bigger:
        small_ind = smaller[0]
        big_ind = bigger[0]
        U[small_ind] = big_ind
        K[big_ind] = K[big_ind] - (1. - K[small_ind])
        if K[big_ind] < 1.:
            bigger.pop(0)
            smaller.append(big_ind)
        smaller.pop(0)
    return K, U

def alias_sample(K, U):
    n = len(K)
    x = np.random.rand()
    i = int(np.floor(n * x))

    if np.random.rand() < K[i]:
        return i
    else:
        return U[i]
```
